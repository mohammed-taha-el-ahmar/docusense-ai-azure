# 02 — Retrieval quality

Exercise the in-memory retriever on the sample corpus. Compare scores for a few queries and inspect the top-k passages.

```python
from pathlib import Path

from docusense.classifier.embeddings import HashedBagOfWordsEmbedding
from docusense.retrieval.chunker import chunk_documents
from docusense.retrieval.indexer import read_corpus
from docusense.retrieval.search import InMemoryHybridRetriever

docs = read_corpus(Path("../data/sample_contracts"))
chunks = chunk_documents(docs, max_tokens=100, overlap_tokens=20)
retriever = InMemoryHybridRetriever(embedding=HashedBagOfWordsEmbedding(dim=256), chunks=chunks)
for query in ["termination for convenience", "net 30 payment", "CPI escalation cap"]:
    print(query)
    for hit in retriever.search(query, top_k=3):
        print(f"  {hit.score:.3f}  {hit.chunk.chunk_id}: {hit.chunk.text[:80]}...")
    print()
```
